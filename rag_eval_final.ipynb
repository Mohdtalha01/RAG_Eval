````python
CORPUS = """
RAG stands for Retrieval-Augmented Generation. It is a technique that combines a
retrieval system with a language model. Instead of relying only on the model's
training data, RAG first retrieves relevant documents from an external database and
then generates an answer grounded in those documents. This reduces hallucinations
and keeps the model's answers up to date.

ChromaDB is an open-source vector database designed for AI applications. It stores
text as embedding vectors — numerical representations that capture semantic meaning.
When you query ChromaDB with a question, it finds the documents whose embeddings are
closest to the question's embedding, even if the exact words don't match. ChromaDB
can run entirely in memory, making it ideal for prototyping.

Faithfulness measures whether the generated answer is supported by the retrieved
context. If your LLM says something that isn't in the retrieved documents, the
score drops. An unfaithful answer contains claims not supported by the retrieved
documents. RAGAS computes faithfulness by asking an evaluator LLM to verify each
claim in the answer against the context.

AnswerRelevancy measures whether the generated answer actually addresses the
question that was asked. A highly faithful answer can still score low on relevancy
if it answers the wrong question. RAGAS measures this by generating reverse
questions from the answer and checking how semantically similar they are to the
original question.

ContextPrecision evaluates whether the retrieved context chunks were actually useful
for answering the question. It compares each retrieved chunk against a reference
answer to judge its relevance. Low precision means your retriever is pulling
irrelevant chunks alongside the useful ones.

ContextRecall checks whether the retriever found all the information needed to fully
answer the question. If the correct answer requires information from three chunks but
your retriever only found two, recall will catch it.

The all-MiniLM-L6-v2 model from HuggingFace is used for generating embeddings. It
runs entirely locally with no API key needed. It is a fast, high-quality model for
semantic similarity and retrieval tasks.

RAGAS v0.4 collections metrics are scored by calling ascore() directly on each
sample. The evaluator LLM is initialized using InstructorLLM with Mode.MD_JSON and
an AsyncOpenAI client pointed at Gemini's OpenAI-compatible endpoint.

Groq is an AI inference company providing fast LLM inference. Their API is free with
generous rate limits. The llama-3.3-70b-versatile model on Groq produces strong
results for RAG generation tasks.
"""
````

````python
from langchain_chroma import Chroma
````

````python
def build_vector_store(texts: list[str]) -> Chroma:
    """
    Splits raw text into chunks and indexes them in an in-memory ChromaDB store.
    Uses all-MiniLM-L6-v2 embeddings — runs locally, no API key needed.
    """
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=50,
    )
    docs = [Document(page_content=t) for t in texts]
    split_docs = splitter.split_documents(docs)

    embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
    return Chroma.from_documents(split_docs, embeddings)
````

````python
def rag_answer(question: str, vector_store: Chroma) -> tuple[str, list[str]]:
    """
    Runs the full RAG pipeline for one question.
    Returns the generated answer AND the retrieved context chunks as plain strings.
    RAGAS needs the contexts separately to evaluate faithfulness and context quality.
    """
    # Retrieve the top 3 most relevant chunks
    retriever = vector_store.as_retriever(search_kwargs={"k": 3})
    retrieved_docs = retriever.invoke(question)

    # RAGAS needs plain strings, not LangChain Document objects
    retrieved_contexts = [doc.page_content for doc in retrieved_docs]

    # Build a prompt that stuffs the retrieved context in
    context_block = "\n\n".join(retrieved_contexts)
    prompt = f"""You are a helpful assistant. Answer the question using ONLY
the information in the context below. If the answer is not in the context,
say "I don't know."

Context:
{context_block}

Question: {question}

Answer:"""

    llm = ChatGroq(model="llama-3.3-70b-versatile")
    response = llm.invoke(prompt)

    return response.content, retrieved_contexts
````

````python
def collect_rag_outputs(vector_store: Chroma) -> list[dict]:
    """
    Runs the RAG pipeline on all test questions.
    Returns a list of dicts containing the question, generated answer,
    retrieved context chunks, and the reference answer.
    RAGAS needs all four to score every metric.
    """
    results = []
    for i, item in enumerate(TEST_QUESTIONS):
        question = item["question"]
        print(f"[{i+1}/{len(TEST_QUESTIONS)}] {question[:65]}...")

        answer, contexts = rag_answer(question, vector_store)
        results.append({
            "question":  question,
            "answer":    answer,
            "contexts":  contexts,
            "reference": item["reference"],
        })
        print(f"    → {answer[:80]}...\n")

    return results
````

````python
def build_evaluation_dataset(rag_outputs: list[dict]) -> EvaluationDataset:
    """
    Converts RAG outputs into a RAGAS EvaluationDataset.
    The field names change slightly:
        question  → user_input
        answer    → response
        contexts  → retrieved_contexts
        reference → reference   (stays the same)
    """
    samples = [
        SingleTurnSample(
            user_input=row["question"],
            response=row["answer"],
            retrieved_contexts=row["contexts"],
            reference=row["reference"],
        )
        for row in rag_outputs
    ]
    return EvaluationDataset(samples=samples)
````

````python
def build_evaluator():

    from langchain_huggingface import HuggingFaceEmbeddings

    # FAST evaluator using Groq
    evaluator_llm = ChatGroq(
        model="llama-3.3-70b-versatile",
        temperature=0,
    )

    # Wrap for RAGAS
    ragas_llm = LangchainLLMWrapper(evaluator_llm)

    # Local embeddings
    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    print("Using Groq evaluator + local embeddings")

    return ragas_llm, embeddings
````

````python
def evaluate_rag(dataset, llm, embeddings):

    print("\nRunning RAGAS evaluation...\n")

    result = evaluate(
        dataset=dataset,
        metrics=[
            faithfulness,
            answer_relevancy,
            context_precision,
            context_recall,
        ],
        llm=llm,
        embeddings=embeddings,
        raise_exceptions=False,
        batch_size=1,
    )

    df = result.to_pandas()

    print(df)

    scores = df.mean(numeric_only=True).to_dict()

    return scores
````

````python
def print_score_report(scores: dict) -> None:
    """
    Prints a human-readable score report with interpretation thresholds.

    Score scale (all metrics):  0.0 = worst   1.0 = perfect

    Metric         What a low score tells you            What to fix
    ─────────────  ────────────────────────────────────  ──────────────────────
    Faithfulness   LLM is hallucinating                  Tighten the prompt;
                                                         add "say I don't know"
    AnswerRelevancy Answer doesn't match the question   Check retrieval or
                                                         prompt framing
    ContextPrecision Retriever pulls noisy chunks       Reduce k; re-tune
                                                         chunk size
    ContextRecall  Retriever misses relevant chunks     Increase k; fix
                                                         chunking strategy
    """
    print("\n" + "=" * 60)
    print("  RAGAS EVALUATION REPORT")
    print("=" * 60)

    thresholds = {
    "faithfulness":     (0.85, "hallucination risk"),
    "answer_relevancy": (0.80, "off-topic answers"),
    "context_precision": (0.75, "noisy retrieval"),
    "context_recall":   (0.75, "incomplete retrieval"),
}

    for metric, score in scores.items():
        threshold, problem = thresholds[metric]
        if math.isnan(score):
            status = "❓ NaN (check API keys / rate limits)"
        elif score >= threshold:
            status = "✅ GOOD"
        elif score >= threshold - 0.15:
            status = "⚠️  FAIR — consider tuning"
        else:
            status = f"❌ LOW — fix {problem}"

        bar_len = int(score * 20) if not math.isnan(score) else 0
        bar = "█" * bar_len + "░" * (20 - bar_len)
        score_str = f"{score:.3f}" if not math.isnan(score) else " NaN "
        # print(f"\n  {metric:<20} {score_str}  [{bar}]  {status}")
        pretty_metric = metric.replace("_", " ").title()

        print(f"\n  {pretty_metric:<20} {score_str}  [{bar}]  {status}")

    print("\n" + "=" * 60)
    # overall = avg([s for s in scores.values() if not math.isnan(s)])
    valid_scores = [
    s for s in scores.values()
    if not math.isnan(s)
      ]

    overall = (
          sum(valid_scores) / len(valid_scores)
          if valid_scores else float("nan")
      )
    print(f"  Overall average: {overall:.3f}")
    print("=" * 60)

    print("""
  HOW TO READ THESE SCORES
  ─────────────────────────
  Faithfulness   — Are answers grounded in retrieved context?
                   < 0.85 → LLM is making things up. Add "say I don't know."

  AnswerRelevancy — Does the answer match the question?
                    < 0.80 → Answer is faithful but off-topic. Fix the prompt.

  ContextPrecision — Were retrieved chunks actually useful?
                     < 0.75 → Retriever pulls noise. Lower k or re-chunk.

  ContextRecall   — Did retrieval find everything needed?
                    < 0.75 → Relevant chunks were missed. Increase k or re-chunk.
""")
````

````python
# ─────────────────────────────────────────────────────────────────────────────
# SECTION 10: Synthetic Test Data Generation (Bonus)
# ─────────────────────────────────────────────────────────────────────────────

def generate_synthetic_questions(corpus: str, n: int = 3) -> list[dict]:
    """
    Uses Groq to auto-generate test question-reference pairs from the corpus.
    Useful when you don't have hand-crafted test questions.

    Returns:
        [
            {
                "question": "...",
                "reference": "..."
            }
        ]
    """

    llm = ChatGroq(
        model="llama-3.3-70b-versatile",
        temperature=0
    )

    prompt = f"""
You are building a high-quality evaluation dataset for a RAG system.

Read the following corpus carefully and generate EXACTLY {n}
question-answer pairs.

Rules:
- Questions must be answerable ONLY from the corpus
- Answers must be concise and factual
- Output ONLY valid JSON
- No markdown
- No explanations

Required format:
[
  {{
    "question": "...",
    "reference": "..."
  }}
]

Corpus:
{corpus}
"""

    response = llm.invoke(prompt)

    try:
        text = (
            response.content
            .replace("```json", "")
            .replace("```", "")
            .strip()
        )

        data = json.loads(text)

        if isinstance(data, list):
            return data

        return []

    except Exception as e:
        print(f"Synthetic generation failed: {e}")
        return []


# ─────────────────────────────────────────────────────────────────────────────
# SECTION 11: Main Orchestrator
# ─────────────────────────────────────────────────────────────────────────────

async def main():

    print("=" * 60)
    print("  RAG EVALUATION WITH RAGAS — COMPLETE PIPELINE")
    print("=" * 60)

    # ─────────────────────────────────────────────────────────────────────
    # Step 1: Build Vector Store
    # ─────────────────────────────────────────────────────────────────────

    print("\n[1/5] Building vector store...")

    corpus_paragraphs = [
        p.strip()
        for p in CORPUS.split("\n\n")
        if p.strip()
    ]

    vector_store = build_vector_store(corpus_paragraphs)

    print(f"      Indexed {len(corpus_paragraphs)} paragraphs.\n")

    # ─────────────────────────────────────────────────────────────────────
    # Optional Synthetic Dataset Generation
    # ─────────────────────────────────────────────────────────────────────

    # Uncomment below if you want synthetic questions
    #
    # print("[Optional] Generating synthetic questions...")
    #
    # synthetic = generate_synthetic_questions(CORPUS, n=5)
    #
    # if synthetic:
    #     global TEST_QUESTIONS
    #     TEST_QUESTIONS = synthetic
    #
    #     print(f"Generated {len(TEST_QUESTIONS)} synthetic questions.\n")

    # ─────────────────────────────────────────────────────────────────────
    # Step 2: Run RAG Pipeline
    # ─────────────────────────────────────────────────────────────────────

    print("[2/5] Running RAG pipeline...\n")

    rag_outputs = collect_rag_outputs(vector_store)

    # Save generated outputs
    with open("rag_outputs.json", "w", encoding="utf-8") as f:
        json.dump(
            rag_outputs,
            f,
            indent=2,
            ensure_ascii=False
        )

    print(f"      Saved {len(rag_outputs)} outputs to rag_outputs.json\n")

    # ─────────────────────────────────────────────────────────────────────
    # Step 3: Build Evaluation Dataset
    # ─────────────────────────────────────────────────────────────────────

    print("[3/5] Building RAGAS dataset...")

    dataset = build_evaluation_dataset(rag_outputs)

    print(f"      Dataset contains {len(dataset.samples)} samples.\n")

    # ─────────────────────────────────────────────────────────────────────
    # Step 4: Build Evaluator
    # ─────────────────────────────────────────────────────────────────────

    print("[4/5] Initializing Gemini evaluator...")

    llm, embeddings = build_evaluator()

    print("      Evaluator ready.\n")

    # ─────────────────────────────────────────────────────────────────────
    # Step 5: Evaluate
    # ─────────────────────────────────────────────────────────────────────

    print("[5/5] Running RAGAS evaluation...\n")

    scores = evaluate_rag(dataset, llm, embeddings)

    # Save scores
    with open("ragas_scores.json", "w", encoding="utf-8") as f:
        json.dump(
            scores,
            f,
            indent=2
        )

    print("\n      Scores saved to ragas_scores.json")

    # ─────────────────────────────────────────────────────────────────────
    # Final Report
    # ─────────────────────────────────────────────────────────────────────

    print_score_report(scores)


# ─────────────────────────────────────────────────────────────────────────────
# RUN
# ─────────────────────────────────────────────────────────────────────────────

if __name__ == "__main__":

    import asyncio

    asyncio.run(main())
````
